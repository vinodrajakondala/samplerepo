```python
import os
import json

def extract_info(directory, keys):
    """
    Extracts specified information from JSON files within a directory (including subdirectories).
    
    :param directory: The root directory to search within.
    :param keys: A list of keys whose values are to be extracted from the JSON files.
    """
    results = []

    # Walk through all subdirectories of the given directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Check if the file is a JSON file
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r') as f:
                        data = json.load(f)
                        # Assuming the keys exist in the 'test4' part of the JSON structure
                        if 'env' in data:
                            extracted_data = {key: data['test4'].get(key, 'N/A') for key in keys}
                            extracted_data['file'] = file_path  # Include file path in the results
                            results.append(extracted_data)
                except json.JSONDecodeError as e:
                    print(f"Error reading JSON from {file_path}: {e}")

    return results

# Define the directory to search and the keys to extract
directory = 'deployment_configurations/preprod/consumers'
keys_to_extract = ['KAFKA_CONSUMER_GROUP_ID', 'KAFKA_TOPICS']

# Extract the information
info = extract_info(directory, keys_to_extract)

# Print or process the extracted information
for item in info:
    print(item)
```

```python
import os
import json

# Define the function to extract specific information from JSON files in specified directories
def extract_info_from_json(root_dir, keys_to_extract):
    """
    Extract specified information from JSON files within specific subdirectories under a given root directory.
    
    :param root_dir: The root directory containing the subdirectories with JSON files.
    :param keys_to_extract: List of keys to extract from the JSON files.
    :return: A list of dictionaries with the extracted information and file paths.
    """
    extracted_info = []
    subdirs = ['preprod/consumers', 'prod/consumers', 'nonprod/consumers']  # Subdirectories to search within

    for subdir in subdirs:
        path = os.path.join(root_dir, subdir)
        for root, dirs, files in os.walk(path):
            for file in files:
                if file.endswith('.json'):
                    file_path = os.path.join(root, file)
                    try:
                        with open(file_path, 'r') as json_file:
                            data = json.load(json_file)
                            # Extract information if present in 'test4'
                            if 'env' in data and isinstance(data['env'], dict):
                                info = {key: data['env'].get(key, 'N/A') for key in keys_to_extract[0:3]}
                                info['file_path'] = file_path  # Include the file path for reference
                                extracted_info.append(info)
                    except json.JSONDecodeError as e:
                        print(f"Error reading JSON from {file_path}: {e}")

    return extracted_info

# Define the directory and keys to extract
directory = "deployment_configurations/"  # Update this path to your specific directory
keys = ['KAFKA_CONSUMER_GROUP_ID', 'KAFKA_TOPICS']

# Extract the information
info_extracted = extract_info_from_json(directory, keys)

# Display the extracted information
for info in info_extracted:
    print(info)
```

```python
# Adjusting the script to extract only up to two KAFKA_TOPICS names if present

def extract_info_from_json_with_limit(root_dir, keys_to_extract, limit_topics=2):
    """
    Extract specified information from JSON files within specific subdirectories under a given root directory,
    limiting the number of KAFKA_TOPICS names if present.
    
    :param root_dir: The root directory containing the subdirectories with JSON files.
    :param keys_to_extract: List of keys to extract from the JSON files, excluding KAFKA_TOPICS.
    :param limit_topics: Maximum number of KAFKA_TOPICS names to extract.
    :return: A list of dictionaries with the extracted information and file paths.
    """
    extracted_info = []
    subdirs = ['preprod/consumers', 'prod/consumers', 'nonprod/consumers']  # Subdirectories to search within

    for subdir in subdirs:
        path = os.path.join(root_dir, subdir)
        for root, dirs, files in os.walk(path):
            for file in files:
                if file.endswith('.json'):
                    file_path = os.path.join(root, file)
                    try:
                        with open(file_path, 'r') as json_file:
                            data = json.load(json_file)
                            # Extract information if present in 'test4'
                            if 'env' in data and isinstance(data['env'], dict):
                                info = {key: data['env'].get(key, 'N/A') for key in keys_to_extract}
                                # Special handling for KAFKA_TOPICS to limit the number of topics
                                kafka_topics = data['env'].get('KAFKA_TOPICS', '').split(',')
                                # Extract up to the limit specified
                                info['KAFKA_TOPICS'] = ','.join(kafka_topics[:limit_topics])
                                info['file_path'] = file_path  # Include the file path for reference
                                extracted_info.append(info)
                    except json.JSONDecodeError as e:
                        print(f"Error reading JSON from {file_path}: {e}")

    return extracted_info

# Define the directory and keys to extract (excluding KAFKA_TOPICS for special handling)
directory = "deployment_configurations"  # Update this path to your specific directory
keys = ['KAFKA_CONSUMER_GROUP_ID']  # KAFKA_TOPICS will be handled separately

# Extract the information with a limit on KAFKA_TOPICS names
info_extracted = extract_info_from_json_with_limit(directory, keys)

# # Display the extracted information
# for info in info_extracted:
#     print(info)
    
# Adjusting the script to print output in a more readable format as requested

def print_info_in_readable_format(info_list):
    """
    Prints the extracted information in a more readable and formatted manner.
    
    :param info_list: List of dictionaries containing the extracted information.
    """
    for info in info_list:
        # Simplifying the file_path to just the filename for readability
        simple_file_path = info['file_path'].split('/')[-1]  # Adjusting for UNIX-like paths
        if '\\' in info['file_path']:  # Adjusting for Windows paths
            simple_file_path = info['file_path'].split('\\')[-1]
        
        # Printing in the desired format
        print("{")
        print(f"  'KAFKA_CONSUMER_GROUP_ID': '{info['KAFKA_CONSUMER_GROUP_ID']}',")
        print(f"  'KAFKA_TOPICS': '{info['KAFKA_TOPICS']}',")
        print(f"  'file_path': '{simple_file_path}'")
        print("}")

# Print the extracted information in a more readable format
print_info_in_readable_format(info_extracted)
```

```python
import json
import os

def create_pafaka_topic_mapping_file(info_list, output_file_path):
    """
    Create a JSON file mapping "pafaka_topic" to "json_file name" for each item in the info_list.

    :param info_list: List of dictionaries containing the extracted information.
    :param output_file_path: The file path where the JSON output will be saved.
    """
    topic_file_mapping = {}
    for info in info_list:
        # Extract the file name from the file path
        json_file_name = os.path.basename(info['file_path'])
        # Split KAFKA_TOPICS by comma in case there are multiple topics
        kafka_topics = info['KAFKA_TOPICS'].split(',')
        for topic in kafka_topics:
            if topic:  # Ensure the topic name is not empty
                # Map the topic to the JSON file name, assuming pafaka_topic is the same as KAFKA_TOPICS
                topic_file_mapping[topic] = json_file_name

    # Save the mapping to a JSON file
    with open(output_file_path, 'w') as outfile:
        json.dump(topic_file_mapping, outfile, indent=4)

# Example usage:
output_path = "pafaka_topic_mapping.json"  # Define the output file path
# Assuming info_extracted is already defined and filled with the extracted information
create_pafaka_topic_mapping_file(info_extracted, output_path)

print(f"Mapping file created at: {output_path}")
```

```python
import csv

def save_info_to_csv(info_list, csv_file_path):
    """
    Saves the extracted information into a CSV file in a readable format.
    
    :param info_list: List of dictionaries containing the extracted information.
    :param csv_file_path: Path to the CSV file where the information will be saved.
    """
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        # Writing the header
        writer.writerow(['KAFKA_CONSUMER_GROUP_ID', 'KAFKA_TOPICS', 'file_path'])
        
        for info in info_list:
            # Simplifying the file_path to just the filename for readability
            simple_file_path = info['file_path'].split('/')[-1]  # Adjusting for UNIX-like paths
            if '\\' in info['file_path']:  # Adjusting for Windows paths
                simple_file_path = info['file_path'].split('\\')[-1]
                
            # Writing the data rows
            writer.writerow([info['KAFKA_CONSUMER_GROUP_ID'], info['KAFKA_TOPICS'], simple_file_path])

# # Assuming info_extracted is your list of dictionaries with the extracted information
# info_extracted = [{
#     'KAFKA_CONSUMER_GROUP_ID': 'nonprod-zoom-events-1', 
#     'KAFKA_TOPICS': 'growthnonprod-zoom-webinar_participant_data,growthnonprod-zoom-meeting_participant_data', 
#     'file_path': 'deployment_configurations\\nonprod/consumers\\skdjfsakj-dsfsd-events-1.json'
# }]

# Define the path to the CSV file where the information will be saved
csv_file_path = 'kafka_info_extracted.csv'

# Save the extracted information to a CSV file
save_info_to_csv(info_extracted, csv_file_path)

# Provide the link to the saved CSV file
csv_file_path
```

```python
def topic_to_json_file(topic_names):
    """
    Converts a list of topic names to their corresponding JSON file names.

    :param topic_names: A list of topic names.
    :return: A list of corresponding JSON file names.
    """
    json_file_names = []

    for topic in topic_names:
        # Extract the country code from the topic name
        parts = topic.split('_')  # Splitting by underscore
        if len(parts) >= 3:  # Ensure there are enough parts to extract the country code
            country_code = parts[2][:2]  # Assuming the country code is the first two characters of the third part
            # Construct the JSON file name
            json_file_name = f"prod-kcr-{country_code.lower()}.json"
            json_file_names.append(json_file_name)

    return json_file_names

# Example usage
topic_names = [
    "dataservices-kvr_prod_au-prod",
    "dataservices-kvr_prod_be-prod",
    "dataservices-kvr_prod_br-prod",
    "dataservices-kvr_prod_ca-prod",
    "dataservices-kvr_prod_ch-prod"
]

json_file_names = topic_to_json_file(topic_names)
for json_file in json_file_names:
    print(json_file)
```
